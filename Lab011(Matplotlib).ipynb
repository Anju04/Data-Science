```python
a1 = [5,8,10]
b1 = [12,16,6]

a2 = [6,9,11]
b2 = [6,15,7]

plt.plot(a1, b1, 'r', label = 'line 1', linewidth = 10)
plt.plot(a2, b2, 'b', label = 'line 2', linewidth = 4)

plt.title('Epic Info', loc='center')  #graph title
plt.ylabel('Y axis')    #Y axis title
plt.xlabel('X axis')    #X axis title

plt.legend(loc='lower left')
plt.grid( color = 'k')
#plt.show()
```

```python
#Histograms
#Histogram is important to see distribution of the data
import numpy as np

data=np.random.randint(0,100,50)
print(data)
plt.hist(data);
```

```python
b=[0,30,50,70,100]
data=np.random.randint(0,100,50)
plt.hist(data,bins=b,color='green',edgecolor='blue') ;
```

```python
plt.hist(data,bins=20,color='green',edgecolor='blue') ;
```

```python
plt.hist(data,bins=10,histtype='bar',color='blue',alpha=0.3);
```
