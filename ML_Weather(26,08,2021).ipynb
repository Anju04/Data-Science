```python
df1=df
```

```python
# Perform normalization

from sklearn.preprocessing import MinMaxScaler
m=MinMaxScaler()
m.fit_transform(df1)
```

```python
m=m.fit_transform(df1)
```

```python
df.columns
```

```python
df2=pd.DataFrame
df2=pd.DataFrame(m,columns=['Temperature_c', 'Humidity', 'Wind_Speed_kmh', 'Wind_Bearing_degrees','Visibility_km', 'Pressure_millibars', 'Rain', 'Description'])
```

```python
df2.head()
```

```python
df2.corr()
```

```python
# Split data into target and features

x=df2.drop('Temperature_c',axis=1)
y=df2.Temperature_c
```

```python
x.head()
```

```python
y.head()
```

```python
# Split data into training and testing

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
# Making a model and training it

from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(x_train,y_train)
```

```python
# Extract Intercept Value
intercept = model.intercept_
print(intercept)
# Extract Coefficient Value
coefficients = model.coef_
print(coefficients)
```

```python
# Genrate Prediction on TestData
predictions = model.predict(x_test)
# length of predictions is 3000 which is equal to length of test cases
```

```python
# Cost functions

from sklearn import metrics

print('MAE', metrics.mean_absolute_error(y_test,predictions))
print('MSE', metrics.mean_squared_error(y_test,predictions))
print('RMSE', np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```

```python
print(metrics.r2_score(y_test,predictions))
```

```python
print(metrics.r2_score)
```
